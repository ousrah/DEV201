```python
from pymongo import MongoClient
client =    MongoClient(host="localhost", port=27017)
print(client)
db = client["dev201"]
print(db)
```

```python
r = db.restaurants.find({"borough":"Manhattan"},{"_id":0,"name":1})
for i in r[:]:
    print(i["name"])
```

```python
selection={}
selection["year"]=1986
r = db.movies.find(selection)
for i in r[:]:
    print(i['title'])
  
```

```python
selection={}
selection["year"]=1986
selection["countries"]="France"
r = db.movies.find(selection)
print(len(list(r)), " movie(s) found")
r.rewind()
for i in r[:]:
    print(i['title'])
```

```python
selection={"year":{"$gt":1980}}
r = db.movies.find(selection)
print(len(list(r)), " movie(s) found")
r.rewind()
for i in r[:]:
    print(i['title'])
```

```python

projection={"_id","title"}
#selection={ "title":{"$regex": "Kate" }}
#selection={ "title":{"$regex": "^Kate" }}
selection={ "title":{"$regex": "Kate$" }}
r = db.movies.find(selection, projection)
for i in r[:]:
    print(i)
```

```python

projection={"_id","title"}
selection={ "title":{"$regex": "Kate$" }}
r = db.movies.find(selection, projection).sort("title",1)
for i in r[:]:
    print(i)
```
